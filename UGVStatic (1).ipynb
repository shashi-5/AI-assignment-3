```python
import heapq
import random

SIZE = 20   

def generate_grid(density):
    grid = []
    for i in range(SIZE):
        row = []
        for j in range(SIZE):
            if random.random() < density:
                row.append(1)  # obstacle
            else:
                row.append(0)
        grid.append(row)
    return grid

def heuristic(a, b):
    return abs(a[0] - b[0]) + abs(a[1] - b[1])

def astar(grid, start, goal):
    pq = []
    heapq.heappush(pq, (0, start))

    came_from = {}
    cost_so_far = {start: 0}

    directions = [(0,1),(1,0),(0,-1),(-1,0)]

    while pq:
        _, current = heapq.heappop(pq)

        if current == goal:
            break

        for d in directions:
            next_node = (current[0]+d[0], current[1]+d[1])

            if 0 <= next_node[0] < SIZE and 0 <= next_node[1] < SIZE:
                if grid[next_node[0]][next_node[1]] == 1:
                    continue

                new_cost = cost_so_far[current] + 1

                if next_node not in cost_so_far or new_cost < cost_so_far[next_node]:
                    cost_so_far[next_node] = new_cost
                    priority = new_cost + heuristic(goal, next_node)
                    heapq.heappush(pq, (priority, next_node))
                    came_from[next_node] = current

    return came_from

grid = generate_grid(0.2)  
start = (0, 0)
goal = (19, 19)

path = astar(grid, start, goal)
print("Path found:", goal in path)
```
