```python
import heapq
import random
import time

SIZE = 20   # change to 70 if needed

# Generate grid with obstacle density
def generate_grid(density):
    grid = []
    for i in range(SIZE):
        row = []
        for j in range(SIZE):
            if random.random() < density:
                row.append(1)  # obstacle
            else:
                row.append(0)
        grid.append(row)
    return grid

# Heuristic (Manhattan distance)
def heuristic(a, b):
    return abs(a[0]-b[0]) + abs(a[1]-b[1])

# A* with Measures of Effectiveness
def astar(grid, start, goal):
    start_time = time.time()

    pq = [(0, start)]
    came_from = {}
    cost_so_far = {start: 0}
    nodes_expanded = 0

    directions = [(0,1),(1,0),(0,-1),(-1,0)]

    while pq:
        _, current = heapq.heappop(pq)
        nodes_expanded += 1

        if current == goal:
            break

        for d in directions:
            nxt = (current[0]+d[0], current[1]+d[1])

            if 0 <= nxt[0] < SIZE and 0 <= nxt[1] < SIZE:
                if grid[nxt[0]][nxt[1]] == 1:
                    continue

                new_cost = cost_so_far[current] + 1

                if nxt not in cost_so_far or new_cost < cost_so_far[nxt]:
                    cost_so_far[nxt] = new_cost
                    priority = new_cost + heuristic(goal, nxt)
                    heapq.heappush(pq, (priority, nxt))
                    came_from[nxt] = current

    end_time = time.time()

    # Path reconstruction
    path = []
    if goal in came_from:
        node = goal
        while node != start:
            path.append(node)
            node = came_from[node]
        path.append(start)
        path.reverse()

    return {
        "success": goal in came_from,
        "path_length": len(path),
        "nodes_expanded": nodes_expanded,
        "time_taken": end_time - start_time
    }


start = (0, 0)
goal = (SIZE-1, SIZE-1)

for density in [0.1, 0.2, 0.3]:
    print("\n--- Obstacle Density:", density, "---")

    grid = generate_grid(density)
    result = astar(grid, start, goal)

    print("Success:", result["success"])
    print("Path Length:", result["path_length"])
    print("Nodes Expanded:", result["nodes_expanded"])
    print("Time Taken:", result["time_taken"], "seconds")
```
